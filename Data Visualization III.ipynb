```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dataset = sb.load_dataset("iris")
dataset
```

```python
fig, axes = plt.subplots(2,2,figsize = (16,9))
sb.histplot(dataset['sepal_length'], ax = axes[0,0])
sb.histplot(dataset['sepal_width'], ax = axes[0,1])
sb.histplot(dataset['petal_length'], ax = axes[1,0])
sb.histplot(dataset['sepal_width'], ax = axes[1,1])
```

```python
fig, axes = plt.subplots(2,2,figsize = (16,9))
sb.boxplot(y = 'petal_length', x = 'species', data = dataset, ax = axes[0,0])
sb.boxplot(y = 'petal_width', x = 'species', data = dataset, ax = axes[0,1])
sb.boxplot(y = 'sepal_length', x = 'species', data = dataset, ax = axes[1,0])
sb.boxplot(y = 'sepal_width', x = 'species', data = dataset, ax = axes[1,1])
```
